```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
p = np.arange(-5,5,0.05)
```

```python
dx, dy = np.meshgrid(p, p)
```

```python
z = (np.sin(dx) + np.sin(dy))
```

```python
plt.imshow(z)
```

```python
plt.imshow(z)
plt.colorbar()
plt.title("Plot for sinX + sinY")
```

```python
#numpy where
a = np.array([1,2,3,4])
b = np.array([100,200,300,400])
check = np.array([True, True, False, False])
```

```python
ans = [(a_val if check else b_val) for a_val,b_val,check in zip(a,b,check)]
```

```python
ans 
```

```python
#same logic with numpy
ans2 = np.where(check, a, b) 
ans2 
```
