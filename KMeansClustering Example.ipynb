```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_blobs
```

```python
d = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
d[0].shape
```

```python
plt.scatter(d[0][:,0], d[0][:,1], c= d[1],cmap='viridis')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=6)
```

```python
kmeans.fit(d[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, sharey = True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(d[0][:,0], d[0][:,1], c=kmeans.labels_)

ax2.set_title('Original')
ax2.scatter(d[0][:,0], d[0][:,1], c=d[1], cmap = 'viridis')
```
