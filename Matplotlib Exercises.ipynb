```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig, axes = plt.subplots(nrows = 1, ncols=2) # axes is and array of objects and each can be accessed individually
# axes.plot(x, y)
axes[0].set_title('First Plot')
axes[0].plot(x,y)
```

```python
fig = plt.figure(figsize = (3,2), dpi = 100) # height 3 inches and width 2 inches
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,'yellow')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('My Figure Plot')
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (8,2))
axes[0].plot(x,y,'cyan')
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_matplotlib_pic.png')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label = 'X Squared')
ax.plot(x, x**3, label = 'X Cubed')
ax.legend(loc=0) # change loc number from 0 - 10 and test it out
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = '#FF8C02', lw = 7, alpha = 0.5, linestyle = "dashed") #alpha for transparency
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='magenta', marker='s', markersize= 10, 
        markerfacecolor = 'blue', markeredgewidth = 3, markeredgecolor = "pink") # to plot the actual values of x on the grapgh
ax.set_xlim([0,3]) # plot zoomed in points
ax.set_ylim([0,5])
```
