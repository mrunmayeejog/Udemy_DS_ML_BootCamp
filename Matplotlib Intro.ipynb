```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#Functional method
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1) # args - no. of rows ie 1, columns 2 and plot number your are refering to
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented method
fig = plt.figure() # we create a blank canvas


ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x,y)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('Set Title')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.4,0.6,0.8,0.8])
ax2 = fig.add_axes([0.1,0.1,0.6,0.6])
ax1.plot(x,y,'g')
ax2.plot(y,x,'pink')
ax1.set_title("Larger Plot")
ax2.set_title("Smaller Plot")
```
