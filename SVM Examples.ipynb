```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
df = pd.DataFrame(cancer['data'], columns = cancer['feature_names'])
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.heatmap(df.corr())
```

```python
from sklearn.model_selection import train_test_split
x= df
y = cancer['target']
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state=101)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, pred))
print(classification_report(y_test, pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# large c values gives to the low bias and high variance. 
# Small C gives more bias less vairiance. So allow more misclassification.
# That it penalizes the misclassification.  
# If gamma is large means the model has low variance
param_grid = {'C':[0.1,1,10,100,1000],'gamma' :[1,0.1,0.01,0.001,0.0001]}
```

```python
grid = GridSearchCV(SVC(), param_grid, verbose=3) # higher the number more verbose
```

```python
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
grid_pred = grid.predict(x_test)
```

```python
print(confusion_matrix(y_test, grid_pred))
print(classification_report(y_test, grid_pred))
```
